### Desafios:

> Ordenar o DataFrame para que na primeira linha tenha a linha com maior gasto, e na última linha o valor com menor gasto (Ordenação).

> Adicionar uma coluna com a região de cada estado

> Adicionar seu estado no DataFrame

```python
df_sem_nulos = df.dropna(axis=1)
```

```python
coluna_total = df_sem_nulos['Total']
```

```python
df_sn = df_sem_nulos.drop(columns=['Total'])
```

```python
df_sn.insert(0, 'Total', coluna_total)
```

```python
df_sn = df_sn.sort_values(by='Total', ascending=False)
```

```python
regioes = {
    '35 São Paulo': 'Sudeste',
    '31 Minas Gerais': 'Sudeste',
    '41 Paraná': 'Sul',
    '43 Rio Grande do Sul': 'Sul',
    '33 Rio de Janeiro': 'Sudeste',
    '29 Bahia': 'Nordeste',
    '26 Pernambuco': 'Nordeste',
    '42 Santa Catarina': 'Sul',
    '23 Ceará': 'Nordeste',
    '52 Goiás': 'Centro-Oeste',
    '15 Pará': 'Norte',
    '21 Maranhão': 'Nordeste',
    '32 Espírito Santo': 'Sudeste',
    '24 Rio Grande do Norte': 'Nordeste',
    '25 Paraíba': 'Nordeste',
    '53 Distrito Federal': 'Centro-Oeste',
    '50 Mato Grosso do Sul': 'Centro-Oeste',
    '22 Piauí': 'Nordeste',
    '27 Alagoas': 'Nordeste',
    '51 Mato Grosso': 'Centro-Oeste',
    '13 Amazonas': 'Norte',
    '28 Sergipe': 'Nordeste',
    '17 Tocantins': 'Norte',
    '11 Rondônia': 'Norte',
    '12 Acre': 'Norte',
    '14 Roraima': 'Norte',
    '16 Amapá': 'Norte'
}
```

```python
df_sn['Regiao'] = df_sn['Unidade da Federação'].map(regioes)
```
