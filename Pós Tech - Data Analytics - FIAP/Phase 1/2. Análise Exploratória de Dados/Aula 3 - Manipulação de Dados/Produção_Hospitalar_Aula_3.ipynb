### Aula 1.2 - Aquisição e Leitura de Dados Reais ll

```python
# Na variável df_numeric, seleciona toda as colunas que são diferentes de "Unidade da Federação",
# e muda os tipos delas para float e os valores nelas que forem string, são representados por NaN.
df_numeric = df.loc[:, df.columns != 'Unidade da Federação'].apply(pd.to_numeric, errors='coerce')
```

```python
# Concatena a coluna "Unidade da Federação" com a váriavel df_numerica, colocando-a na posição 1.
df = pd.concat([df['Unidade da Federação'], df_numeric], axis=1)
```

```python
df.head()
```

```python
# Pegamos todas as colunas e tiramos a média delas com excessão da primeira,
# que é uma string.
media_columns = df.iloc[:, 1:].mean()
```

### Desafio 2 - Deixar o eixo x em 45 graus ou 0 grau para facilitar a leitura dos estados.

```python
import numpy as np

# Criação do gráfico
fig, ax = plt.subplots(figsize=(15, 6))
bars = ax.barh(df['Unidade da Federação'], df['2021/Set'], color='skyblue')

# Adicionando os valores dentro das barras
for bar in bars:
    xval = bar.get_width()
    if np.isfinite(xval):
        ax.text(xval, bar.get_y() + bar.get_height() / 2, f'{xval:,.2f}',
                ha='left', va='center')

# Configurações do gráfico
ax.set_title('Valor por Unidade da Federação')
ax.set_ylabel('Unidade da Federação')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Remove os valores do eixo x
ax.set_xticks([])

plt.show()
```
