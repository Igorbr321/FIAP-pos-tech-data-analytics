```python
# Na variável df_numeric, seleciona toda as colunas que são diferentes de "Unidade da Federação",
# e muda os tipos delas para float e os valores nelas que forem string, são representados por NaN.
df_numeric = df.loc[:, df.columns != 'Unidade da Federação'].apply(pd.to_numeric, errors='coerce')
```

```python
# Concatena a coluna "Unidade da Federação" com a váriavel df_numerica, colocando-a na posição 1.
df = pd.concat([df['Unidade da Federação'], df_numeric], axis=1)
```

### Desafio 2 - Deixar o eixo x em 45 graus ou 0 grau para facilitar a leitura dos estados.

```python
import numpy as np

# Criação do gráfico
fig, ax = plt.subplots(figsize=(15, 6))
bars = ax.barh(df['Unidade da Federação'], df['2021/Set'], color='skyblue')

# Adicionando os valores dentro das barras
for bar in bars:
    xval = bar.get_width()
    if np.isfinite(xval):
        ax.text(xval, bar.get_y() + bar.get_height() / 2, f'{xval:,.2f}',
                ha='left', va='center')

# Configurações do gráfico
ax.set_title('Valor por Unidade da Federação')
ax.set_ylabel('Unidade da Federação')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Remove os valores do eixo x
ax.set_xticks([])

plt.show()
```

### Desafios:

> Ordenar o DataFrame para que na primeira linha tenha a linha com maior gasto, e na última linha o valor com menor gasto (Ordenação).

> Adicionar uma coluna com a região de cada estado

> Adicionar seu estado no DataFrame

```python
df_sn = df.dropna(axis=1)
```

```python
coluna_total = df_sn['Total']
```

```python
df_sn = df_sn.drop(columns=['Total'])
```

```python
df_sn.insert(0, 'Total', coluna_total)
```

```python
df_sn = df_sn.sort_values(by='Total', ascending=False)
```

```python
regioes = {
    'AC': 'Norte', 'AL': 'Nordeste', 'AM': 'Norte', 'AP': 'Norte', 'BA': 'Nordeste',
    'CE': 'Nordeste', 'DF': 'Centro-Oeste', 'ES': 'Sudeste', 'GO': 'Centro-Oeste',
    'MA': 'Nordeste', 'MT': 'Centro-Oeste', 'MS': 'Centro-Oeste', 'MG': 'Sudeste',
    'PA': 'Norte', 'PB': 'Nordeste', 'PR': 'Sul', 'PE': 'Nordeste', 'PI': 'Nordeste',
    'RJ': 'Sudeste', 'RN': 'Nordeste', 'RS': 'Sul', 'RO': 'Norte', 'RR': 'Norte',
    'SC': 'Sul', 'SP': 'Sudeste', 'SE': 'Nordeste', 'TO': 'Norte'
}
```

```python
def adicionar_regiao(estado):
    codigo = estado.split(' ')[0]  # Pega apenas o código do estado
    regiao = regioes.get(codigo[:2])  # A primeira parte do código é o código do estado
    return regiao
```

```python
df_sn['Regiao'] = df_sn['Unidade da Federação'].apply(adicionar_regiao)
```
