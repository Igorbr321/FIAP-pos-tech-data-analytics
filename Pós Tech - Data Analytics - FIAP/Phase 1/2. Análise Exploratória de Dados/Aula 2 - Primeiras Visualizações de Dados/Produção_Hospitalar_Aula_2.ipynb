## Pré-processamento de Dados

```python
# Importando biblioteca pandas.
import pandas as pd
```

```python
# Deixando um formato padrão de 2 casas decimais.
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Carregando o dataframe, que não está no formato utf-8.
# Usando funções, modificamos para o tipo dos dados do dataframe e excluimos algumas linhas, que não estão em formato tabular.
file = 'A152918189_28_143_208.csv'

df = pd.read_csv(file, sep=";", encoding="ISO-8859-1", skiprows=3, skipfooter=12, decimal=",", thousands=".", engine="python")
```

## Aula - 1 Aquisição e Leitura de Dados Reais

```python
# Na variável df_numeric, seleciona toda as colunas que são diferentes de "Unidade da Federação",
# e muda os tipos delas para float e os valores nelas que forem string, são representados por NaN.
df_numeric = df.loc[:, df.columns != 'Unidade da Federação'].apply(pd.to_numeric, errors='coerce')
```

```python
# Concatena a coluna "Unidade da Federação" com a váriavel df_numerica, colocando-a na posição 1.
df = pd.concat([df['Unidade da Federação'], df_numeric], axis=1)
```

```python
df.head()
```

```python
# Pegamos todas as colunas e tiramos a média delas com excessão da primeira,
# que é uma string.
media_columns = df.iloc[:, 1:].mean()
```

```python
media_columns.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = df.plot(x='Unidade da Federação', y='2008/Ago', kind='bar', figsize=(9, 6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('Valor por Unidade da Federação')
plt.show()
```

### Desafio 1 - Executar o gráfico anterior com o mês mais recente

```python
axis = df.plot(x='Unidade da Federação', y='2021/Set', kind='bar', figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('Valor por Unidade da Federação')
plt.show()
```

### Desafio 2 - Deixar o eixo x em 45 graus ou 0 grau para facilitar a leitura dos estados.

```python
import numpy as np

# Criação do gráfico
fig, ax = plt.subplots(figsize=(15, 6))
bars = ax.barh(df['Unidade da Federação'], df['2021/Set'], color='skyblue')

# Adicionando os valores dentro das barras
for bar in bars:
    xval = bar.get_width()
    if np.isfinite(xval):
        ax.text(xval, bar.get_y() + bar.get_height() / 2, f'{xval:,.2f}',
                ha='left', va='center')

# Configurações do gráfico
ax.set_title('Valor por Unidade da Federação')
ax.set_ylabel('Unidade da Federação')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

# Remove os valores do eixo x
ax.set_xticks([])

plt.show()
```
