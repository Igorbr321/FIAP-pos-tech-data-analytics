```python
df_sem_nulos = df_sem_nulos.drop('Total', axis=1)
```

```python
df_sem_nulos_T = df_sem_nulos.T
```

```python
df_sem_nulos_T.plot(figsize=(12,6))
```

### Desafios:

> Reposicionar a legenda do gráfico

> Retocar título da visualização

> Colocar os títulos nos dois eixos.

```python
from matplotlib.ticker import ScalarFormatter
import matplotlib.ticker as ticker

ax = df_sem_nulos_T.plot(figsize=(12, 6))
plt.title('Unidade da Federação', fontsize=14)
plt.legend(loc='upper left', title='Estados', fontsize=10, frameon=False, bbox_to_anchor=(1.05, 1))
ax.set_ylabel(ylabel='Valor')
ax.set_xlabel(xlabel='Tempo')

ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=False))
ax.ticklabel_format(style='plain', axis='y')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.show()
```
