```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import ScalarFormatter
```

```python
df.replace('-', np.nan, inplace=True)
```

```python
df_sem_nulos = df.dropna(axis=1)
```

```python
df_sem_nulos.head()
```

```python
df_sem_nulos.set_index('Unidade da Federação', inplace=True)
```

```python
df_sem_nulos = df_sem_nulos.sort_values(by='Total', ascending=False)
df_sem_nulos.drop(columns='Total',axis= 1, inplace=True)
```

```python
colunas_interessadas = df_sem_nulos.columns[6:]
```

```python
df_sem_nulos = df_sem_nulos[colunas_interessadas]
```

## Desafios

```python
df_sem_nulos = df_sem_nulos /1_000_000
```

```python
# Criar o gráfico
ax = df_sem_nulos.head(5).T.plot(figsize=(12,6))

# Título do Gráfico
plt.title('Unidade da Federação', fontsize=14, fontweight='bold')

# Rótulo dos eixos
ax.set_xlabel('Datas', fontsize=12, fontweight='bold')
ax.set_ylabel('Valores (por milhão)', fontsize=12, fontweight='bold')

# Legenda
plt.legend(title='Estados', title_fontsize=12, labelspacing=1.2, loc='upper left', fontsize=10, frameon=False, bbox_to_anchor=(1.05, 1))

# Configurações do Eixo y
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.show()
```

A função `sort_index` é usada para ordenar os dados em um DataFrame ou Series com base nos índices. É particularmente útil quando você deseja reorganizar os dados por seus rótulos de índice (linhas ou colunas) em ordem crescente ou decrescente.

**Sintaxe**:  `DataFrame.sort_index(axis=0, level=None, ascending=True, inplace=False, kind='quicksort', na_position='last', sort_remaining=True, ignore_index=False, key=None)`
